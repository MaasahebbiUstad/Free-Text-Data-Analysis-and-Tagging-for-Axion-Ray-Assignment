# Data Cleaning

```python
# Iterating over each column in the DataFrame to deal with null values

for i in df:
    if df[i].dtype == "O":  # Checking for object type (string or categorical)
        df[i] = df[i].fillna(df[i].mode()[0])  # Reassigning the filled column
    elif df[i].dtype in ("int","float"):  # Checking for integer type
        df[i] = df[i].fillna(df[i].median())  # Reassigning the filled column
```

```python
# Counting and sorting the number of missing (null) values in each column, in descending order

df.isnull().sum().sort_values(ascending=False)
```

```python
# Print unique values (sorted) for each column along with their data type for inspection

for i in df:
  print(i,df[i].astype(str).sort_values().unique(),"\n",df[i].dtypes)
  print()
  print("-----------------------------------------------------------------------------")
```

```python
# Replace specific noise values in 'CUSTOMER_VERBATIM' and 'CORRECTION_VERBATIM' columns with NaN

df['CUSTOMER_VERBATIM'] = df['CUSTOMER_VERBATIM'].replace('�������� ������ ����������', np.nan, regex=False)

df['CORRECTION_VERBATIM'] = df['CORRECTION_VERBATIM'].replace('方向盘底部的皮革脱落了。拆下方向盘并更换新的。CC：0890 FC：2039PRA#490428700000 人工 OP：0130 0.50 人工', np.nan)

```

```python
# Dropped 'CAMPAIGN_NBR' (empty column), 'TRANSACTION_CNTR' and 'COMPLAINT_CD_CSI' for data cleaning

df=df.drop(['CAMPAIGN_NBR','TRANSACTION_CNTR','COMPLAINT_CD_CSI'],axis=1 )
```

**TRANSACTION_CNTR (only contains 0s):**

This column doesn't provide any useful variation because it’s the same value (0) throughout the entire dataset. It doesn't offer any new insights or information for analysis or modeling, so it might be a candidate for removal.

**COMPLAINT_CD_CSI (only contains 1s):**

Similar to the above, this column doesn't vary and only contains 1. It may not be useful unless it serves as a flag or indicator, but in most cases, it’s also a constant.

# Visualization:

Frequency Distribution of a Categorical Column:

```python
# Iterating over each column in the DataFrame, checking for object type columns with less than 6 unique values

for i in df:
  if df[i].dtypes=="O":
    if df[i].nunique()<6:
      print(df[i].value_counts())
      sorted_categories = df[i].value_counts().index
      sns.countplot(data=df,x=df[i],palette='coolwarm',order=sorted_categories)
      plt.show()
      print()
      print("-------------------------------------------------------------------------------------------------------")
```

**# Histogram for a numerical column**

```python
# Generate and display histograms with KDE for each numerical column to visualize the distribution of values

for i in df:
  if df[i].dtypes!="O":
    sns.histplot(df[i], kde=True, color='skyblue')
    plt.title(i)
    plt.show()
    print()
    print( "-------------------------------------------------------------------------------------------------")
```

# **Detect and Handle Outliers:**

```python
# Iterate through each column in the DataFrame, and create a boxplot for non-categorical columns

for i in df.columns:
    if df[i].dtypes != "O":  # Check if the column is not categorical
        plt.figure()
        sns.boxplot(y=df[i])  # Pass the column as the y-axis
        plt.title(f"Boxplot of {i}")
        plt.show()
        print()
        print("---------------------------------------------------------------------------------------")
```

Keeping both columns in the dataset could be unnecessary and redundant, as they are providing identical details.

**Strong Relationship:**

The values in these two columns ('TOTALCOST' and 'REPORTING_COST') are strongly correlated, meaning that they follow a similar pattern and move in the same direction.
A correlation of 0.97 suggests that a large proportion of the variation in one column can be explained by the variation in the other column.
